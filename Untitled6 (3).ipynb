```python
df=pd.get_dummies(data=df,columns=['Manufacturer'],dtype='int')
df
```

```python
x=df.drop('Price',axis=1)
x
```

```python
y=df['Price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression() 
linear_regressor.fit(x_train,y_train)
```

```python
y_pred_lr = linear_regressor.predict(x_test)

y_pred_lr
```

```python
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse


print("----Linear Regresssion - Model Evaluation----")
print("mean Absolute Error(MAE)",mae(y_test,y_pred_lr))
print("mean Squared Error(MSE)",mse(y_test,y_pred_lr))
print("Root Mean Squared Error (RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
                                                
```

```python
print("Accuracy of LR on training set:",linear_regressor.score(x_train,y_train))
print("Accuracy of Lr  on test set:",linear_regressor.score(x_test,y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtc=DecisionTreeRegressor()
dtc.fit(x_train,y_train)
pred_dtc=dtc.predict(x_test)
print("mean absolute error(mae):",mae(y_test,pred_dtc))
print("mean Squared Error(MSE)",mse(y_test,pred_dtc))
print("Root Mean Squared Error (RMSE):",np.sqrt(mse(y_test,pred_dtc)))
```

```python
print("accuracy of dtc on training set:",dtc.score(x_train,y_train))
print("Accuracy of dtc  on test set:",dtc.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
pred_rfr=rfr.predict(x_test)
print("mean absolute error(mae):",mae(y_test,pred_rfr))
print("mean Squared Error(MSE)",mse(y_test,pred_rfr))
print("Root Mean Squared Error (RMSE):",np.sqrt(mse(y_test,pred_rfr)))
```

```python
print("accuracy of rfR on training set:",rfr.score(x_train,y_train))
print("Accuracy of rfr  on test set:",rfr.score(x_test,y_test))
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
pred_svr=svr.predict(x_test)
print("mean absolute error(mae):",mae(y_test,pred_svr))
print("mean Squared Error(MSE)",mse(y_test,pred_svr))
print("Root Mean Squared Error (RMSE):",np.sqrt(mse(y_test,pred_svr)))
```

```python
print("accuracy of svR on training set:",svr.score(x_train,y_train))
print("Accuracy of svr  on test set:",svr.score(x_test,y_test))
```
